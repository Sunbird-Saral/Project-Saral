```python
import uuid
import json

roiIndex = 1
cellIndex = 1

def get_annotation(filename):

    with open(filename) as f:
        data = json.load(f)
        f.close()
        return data['regions']

def get_rois(regions,tagGroup):
    rois  = []
    index = 0
    global roiIndex 
    for region in regions:
        if region['tags'][0].startswith(tagGroup):
            if region['tags'][0].startswith("ROLLNUMBER"):
                extractionValue = "NUMERIC_CLASSIFICATION"
            else:
                extractionValue = "CELL_OMR"
            rois.append({

                "annotationTags": str(region['tags'][0]),
                "extractionMethod": extractionValue,
                "roiId": str(roiIndex),
                "index": index,
                "rect": {

                    "top": int(region['boundingBox']['top']),

                    "left": int(region['boundingBox']['left']),

                    "bottom": int(region['boundingBox']['top']) + int(region['boundingBox']['height']),

                    "right": int(region['boundingBox']['left']) + int(region['boundingBox']['width'])

                }

            })
            index = index + 1
            roiIndex = roiIndex +1 
        
    return rois

def get_cells(regions,tagGroups,formatLookup,formatNameLookup):
    
    cells_data = []
    renderIndex = 1
    global cellIndex
    for tagGroup in tagGroups: 
              try:
                  formatValue = formatLookup[str(tagGroup)]
              except KeyError as ke:
                  formatValue = ""
            
              try:
                  formatName = formatNameLookup[str(tagGroup)]
              except KeyError as ke:
                  formatName = ""

              cells_data.append({
                          "cellId": str(cellIndex),
                          "rois": get_rois(regions,tagGroup),
                          "render": {
                              "index": renderIndex
                          },
                          "format": {
                              "name": formatName,
                              "value": formatValue
                          },
                          "validate": {
                              "regExp": ""
                          }
              })
              renderIndex = renderIndex +1   
              cellIndex = cellIndex +1 
    return cells_data

def get_layout(cells):
    layout_data = []
    layout_data.append({
        "layout": {
            "version": "1.0",
            "name": "ANY1S5QMULTICHOICE Exam Sheet Form",
            "cells": cells
        }
    })    
    return layout_data[0]

def pp_json(json_thing, sort=True, indents=4):
    if type(json_thing) is str:
        print(json.dumps(json.loads(json_thing), sort_keys=sort, indent=indents))
    else:
        print(json.dumps(json_thing, sort_keys=sort, indent=indents))
    return None
```

```python
regions=get_annotation("any_1s_5q_multichoice_vottraw.json")
```

```python
tagGroups = ["ROLLNUMBER", "QUESTION1","QUESTION2","QUESTION3","QUESTION4","QUESTION5"]

formatLookup = {
    'ROLLNUMBER': 'ROLLNUMBER',
    'QUESTION1':'QUESTION 1',
    'QUESTION2':'QUESTION 2',
    'QUESTION3':'QUESTION 3',
    'QUESTION4':'QUESTION 4',
    'QUESTION5':'QUESTION 5',
}

formatNameLookup = {
    'ROLLNUMBER': 'ROLLNUMBER',
    'QUESTION1':'QUESTION1',
    'QUESTION2':'QUESTION2',
    'QUESTION3':'QUESTION3',
    'QUESTION4':'QUESTION4',
    'QUESTION5':'QUESTION5',

}
```

```python
cells=get_cells(regions,tagGroups,formatLookup,formatNameLookup)
```

```python
pp_json(get_layout(cells),False)
```
